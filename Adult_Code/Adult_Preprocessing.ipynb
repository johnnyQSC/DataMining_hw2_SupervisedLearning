1.訓練資料前處理

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Adult_Train.csv")
df
```

```python
df.apply(lambda x:np.sum(x==' ?'))
```

```python
fliter1 = (df["native-country"] != ' ?')
fliter2 = (df["occupation"] != ' ?')
fliter3 = (df["workclass"] != ' ?')
df_1 = df[fliter1 & fliter2 & fliter3]
print(df_1['native-country'].value_counts())
print(df_1['occupation'].value_counts())
print(df_1['workclass'].value_counts())
```

```python
replace_values = {'Preschool':0,'1st-4th':1,'5th-6th':2,'7th-8th':3,'9th':4,'10th' :5,'11th' :6,'12th' : 7,'HS-grad' :8,'Prof-school':9,'Assoc-acdm' :10,'Assoc-voc':11,'Some-college':12,'Bachelors':13 ,'Masters':14,'Doctorate':15,}                                                                                          
df_2=df_1.replace({"education": replace_values})  
print(df_2['education'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df_3= df_2
df_3['workclass'] = labelencoder.fit_transform(df_3['workclass'])
df_3['education']=labelencoder.fit_transform(df_3['education'])
df_3['occupation'] = labelencoder.fit_transform(df_3['occupation'])
df_3['marital-status'] = labelencoder.fit_transform(df_3['marital-status'])
df_3['relationship'] = labelencoder.fit_transform(df_3['relationship'])
df_3['race'] = labelencoder.fit_transform(df_3['race'])
df_3['sex'] = labelencoder.fit_transform(df_3['sex'])
df_3['native-country'] = labelencoder.fit_transform(df_3['native-country'])
df_3['salary'] = labelencoder.fit_transform(df_3['salary'])
df_3
```

```python
df_3.to_csv("Adult_FinalTrain.csv",index=False,encoding="utf-8-sig")
```

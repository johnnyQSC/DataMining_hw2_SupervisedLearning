1.訓練資料集

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Adult_FinalTrain.csv")
df
```

```python
#正規化
from sklearn.preprocessing import MinMaxScaler
z_score=MinMaxScaler()
df_1=z_score.fit_transform(df)
df_2=pd.DataFrame(df_1,columns=df.columns)
df_2
```

```python
df_2.describe()
```

```python
x_train = df_2.loc[:,['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
                      ,'race','sex','capital-gain','capital-loss','native-country','salary']]
y_train = df_2.loc[:,['hours-per-week']]
```

```python
#建立隨機森林模型
from sklearn.ensemble import RandomForestRegressor
randomforest=RandomForestRegressor(n_estimators=1000,
                                   n_jobs=-1,
                                   oob_score = True,
                                   bootstrap = True,
                                   random_state = 42)
randomforest.fit(x_train,y_train)
y_train.predicted=randomforest.predict(x_train)
```

```python
print('R^2 Training Score: {:.2f} \nOOB Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(randomforest.score(x_train, y_train),randomforest.oob_score_,randomforest.score(x_train, y_train)))
```

```python
randomforest.feature_importances_
```

```python
feature=x_train.columns
important=pd.DataFrame(index=feature)
important['Values']=randomforest.feature_importances_
```

```python
important.plot.bar()
```

2.測試資料集

```python
testData=pd.read_csv("Adult_FinalTest.csv")
testData
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#正規化
z_score= MinMaxScaler()
testData2 = z_score.fit_transform(testData)
testData3 = pd.DataFrame(testData2,columns=testData.columns)
testData3
```

```python
testData3.describe()
```

```python
X_train = testData3.loc[:,['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
                      ,'race','sex','capital-gain','capital-loss','native-country','salary']]
Y_train = testData3.loc[:,['hours-per-week']]
```

```python
#建立隨機森林模型
from sklearn.ensemble import RandomForestRegressor
rf_Test=RandomForestRegressor(n_estimators=1000,
                              n_jobs=-1,
                              oob_score = True,
                              bootstrap = True,
                              random_state = 42)
rf_Test.fit(X_train,Y_train)
Test_Y_predicted = rf_Test.predict(X_train)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE: ',mean_absolute_error(Y_train,Test_Y_predicted))
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
print('RMSE： ',np.sqrt(mean_squared_error(Y_train,Test_Y_predicted)))
```

1.訓練

```python
import xgboost as xgb
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Adult_FinalTrain.csv")
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
#正規化
z_score= MinMaxScaler()
df1= z_score.fit_transform(df)
df2=pd.DataFrame(df1,columns=df.columns)
df2
```

```python
x_train = df2.loc[:,['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
                      ,'race','sex','capital-gain','capital-loss','hours-per-week','native-country','salary']]
y_train = df2.loc[:,['hours-per-week']]
```

```python
model=xgb.DMatrix(data=x_train,label=y_train)
```

```python
#建立XGBoost模型
xg_regression=xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_regression.fit(x_train,y_train)
```

```python
y_pred=xg_regression.predict(x_train)
```

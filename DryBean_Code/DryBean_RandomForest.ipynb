```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("DryBean_FinalDataset.csv")
df
```

```python
df['Class'].value_counts()
```

```python
#正規化
from sklearn.preprocessing import MinMaxScaler
z_score=MinMaxScaler()
df_1=z_score.fit_transform(df)
df_2=pd.DataFrame(df_1,columns=df.columns)
df_2
```

```python
df_2.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df_2.loc[:,['Area','Perimeter','MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity',
                    'ConvexArea','EquivDiameter','Extent','Solidity','roundness','Compactness',
                    'ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4','Class']]
y= df_2.loc[:,['Class']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
#建立隨機森林模型
from sklearn.ensemble import RandomForestRegressor
randomforest=RandomForestRegressor(n_estimators=1000,
                                   n_jobs=-1,
                                   oob_score = True,
                                   bootstrap = True,
                                   random_state = 42)
randomforest.fit(x_train,y_train)
y_train.predicted=randomforest.predict(x_train)
```

```python
print('R^2 Training Score: {:.2f} \nOOB Score: {:.2f} \nR^2 Validation Score: {:.2f}'.format(randomforest.score(x_train, y_train),randomforest.oob_score_,randomforest.score(x_train, y_train)))
```

```python
randomforest.feature_importances_
```

```python
feature=x_train.columns
important=pd.DataFrame(index=feature)
important['Values']=randomforest.feature_importances_
```

```python
important.plot.bar()
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE:',mean_absolute_error(y_train,y_train.predicted))
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
print('RMSE：',np.sqrt(mean_squared_error(y_train,y_train.predicted)))
```

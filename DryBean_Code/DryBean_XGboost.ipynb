```python
from sklearn.model_selection import train_test_split
```

```python
x = df2.loc[:,['Area','Perimeter','MajorAxisLength','MinorAxisLength','AspectRation','Eccentricity',
                    'ConvexArea','EquivDiameter','Extent','Solidity','roundness','Compactness',
                    'ShapeFactor1','ShapeFactor2','ShapeFactor3','ShapeFactor4','Class']]
y= df2.loc[:,['Class']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
model=xgb.DMatrix(data=x_train,label=y_train)
```

```python
#建立XGBoost模型
xg_regression=xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,
                max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
xg_regression.fit(x_train,y_train)
```

```python
y_pred=xg_regression.predict(x_train)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
print('MAE:',mean_absolute_error(y_train,y_pred))
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
print('RMSE：',np.sqrt(mean_squared_error(y_train,y_pred)))
```
